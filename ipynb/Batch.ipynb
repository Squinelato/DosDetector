# Computing in chunk the whole dataset

```python
def batchPreprocessing(path, chunksize):

    start_time = perf_counter()
    
    collection = list()
    total_lines = !wc -l $path | grep -Eo '^[0-9]+'
    total_lines = int(total_lines[0]) 

    for i, chunk in enumerate(pd.read_csv(path, low_memory=False, 
                                       index_col=[0], chunksize=chunksize)):

        porcentage = (i * chunksize) / total_lines

        print(round(porcentage * 100, 2), '% complete')

        preprocessed_chunk = preprocessing(chunk, '1s')
        collection.append(preprocessed_chunk)

    print('#'*50, '\nTotal time:', perf_counter() - start_time)
    
    return pd.concat(collection)
```

```python
processed_dataset = batchPreprocessing(path='../dataset/raw/SUEE1.csv', chunksize=100000)
```
